```python
#Feature 1
df["savings_rate"] = (df["income"] - df["monthly_spend"]) / df["income"]
```

Savings Rate (savings_rate)

Formula: (income - monthly_spend) / income

Why useful: Measures how much of their income a person actually saves. Higher savings rate can indicate financial stability and lower risk for lenders. This gives more insight than just raw income or spending alone.
